```python
# 假设 `model` 是你预训练好的模型对象
similar_words = [key for key, value in model.most_similar('ferroelectricity', topn=500)]

import re
# 定义匹配复杂化学式的正则表达式模式
chemical_formula_pattern = re.compile(r'\b[A-Z][a-z]?[A-Za-z0-9_\.]*\b')

# 存储提取的化学式
chemical_formulas = []

# 提取化学式
for word in similar_words:
    matches = chemical_formula_pattern.findall(word)
    for match in matches:
        # 过滤出有意义的化学式，忽略单个字符和数字
        if len(match) > 1 and re.search(r'[A-Za-z]', match):
            chemical_formulas.append(match)

# 去重
chemical_formulas = list(set(chemical_formulas))

# 打印所有提取到的化学式
print(chemical_formulas)
```
